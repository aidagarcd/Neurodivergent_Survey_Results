### Graphics

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Conteo agrupado
counts = survey.groupby(['SEX', 'ARE YOU NEURODIVERGENT']).size().reset_index(name='count')

plt.figure(figsize=(12, 7))  

bar_plot = sns.barplot(x='SEX', y='count', hue='ARE YOU NEURODIVERGENT', data=counts)

# Agregar etiquetas a las barras usando bar_label
ax = bar_plot
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')  # Coloca etiquetas en el borde de las barras

plt.title('Total sex respondents and whether they are neurodivergent')
plt.ylabel('Count')
plt.xlabel('Sex')

plt.show()
```

#### The conclusion from the total number of respondents is that slightly more women than men responded and that there are slightly more neurodivergent women than men among the respondents.

```python
plt.figure(figsize=(12, 7))

sns.countplot(
    x='YOU INVITED SOMEONE WHO BEHAVES DIFFERENTLY IN EVENTS', 
    hue='YOU IGNORE THEM / MAKE YOU UNCOMFORTABLE', 
    data=survey, 
    palette=['green', 'red']  # Colores personalizados
)

plt.title('Social exclusion at events')
plt.xlabel('Invitation to events')
plt.ylabel('Count')
plt.legend(title='Feel uncomfortable')
plt.show()
```

#### There are the same number of people who do not invite neurodiverse people to events as there are people who do invite neurodiverse people, in the case of people who do not invite neurodiverse people they feel more uncomfortable with them than those who do invite them, and despite inviting them there is a percentage of them who feel uncomfortable with the presence of people with atypical behaviour.

```python
survey['MEASURES ARE NEEDED TO INTEGRATE THEM'] = survey['MEASURES ARE NEEDED TO INTEGRATE THEM'].str.strip()
survey['SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.'] = survey['SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.'].str.strip()

# Contar la frecuencia combinada de las categorÃ­as
counts = survey.groupby(['MEASURES ARE NEEDED TO INTEGRATE THEM', 'SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.']).size().reset_index(name='counts')

plt.figure(figsize=(12, 7))
sns.barplot(
    data=counts,
    x='MEASURES ARE NEEDED TO INTEGRATE THEM',
    y='counts',
    hue='SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.',
    palette="Set2"
)
plt.title('Perceptions of integration measures')
plt.xlabel('Measures needed for integration')
plt.ylabel('Count')

# Mover la leyenda a la izquierda
plt.legend(title='Inclusive environments', loc='upper left', bbox_to_anchor=(-0.005, 1))

plt.show()
```

#### The feeling that there are no measures to integrate these people is notorious above the other answers and the majority of people think that the current social spaces are not inclusive with these people.

```python
counts = survey['YOU HAVE WITNESSED INJUSTICE WITH NEUROD.'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',  # percent
    startangle=90,      
    colors=plt.cm.Paired.colors,  
    wedgeprops={'edgecolor': 'black'}  #black width
)

plt.title('Distribution of Witnessed Injustices')
plt.show()
```

#### There is a general perception of injustice in social environments towards neurodiverse people.

```python
comparison_counts = survey.groupby(['YOU HAVE WITNESSED INJUSTICE WITH NEUROD.', 'YOU IGNORE THEM / MAKE YOU UNCOMFORTABLE']).size().reset_index(name='counts')


plt.figure(figsize=(12, 7))
sns.barplot(
    x='YOU HAVE WITNESSED INJUSTICE WITH NEUROD.', 
    y='counts', 
    hue='YOU IGNORE THEM / MAKE YOU UNCOMFORTABLE', 
    data=comparison_counts,
    palette='Set2'
)

plt.title('Comparison: Witnessed Injustice vs Social Discomfort', fontsize=14)
plt.xlabel('Witnessed Injustice', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Social Discomfort', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.show()
```

#### The more social discomfort with them the more injustice against them

```python
cross_tab = pd.crosstab(survey['WE SHOULD HAVE MORE INFO'], survey['SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.'], normalize=True)

cross_tab.plot(kind='bar', stacked=True, colormap="viridis")

plt.title('Relationship between two answers')
plt.xlabel('WE SHOULD HAVE MORE INFO')
plt.ylabel('Counts')
plt.legend(title='Social Environments')
plt.show()
```

#### As we can see there is a clear relationship between the lack of information that people have clearly answered that there is no information about socialisation with neurodiverse people and the fact that it is missed that spaces and social events do not include these people, there are no tips or techniques or anything at all to integrate these people.

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.countplot(x='DID YOU HELPED HIM', hue='YOU KNOW HOW TO HELP', data=survey, palette='viridis')

plt.title("'YOU KNOW HOW TO HELP' vs 'DID YOU HELPED HIM'")
plt.xlabel("You know how to help")
plt.ylabel("Counts")
plt.legend(title="You helped them")
plt.show()
```

#### We can see that people mostly help potential neurodiverse people in the environment but they do not know how to help.

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.countplot(
    x='YOU INVITED SOMEONE WHO BEHAVES DIFFERENTLY IN EVENTS',
    hue='SOCIAL ENVIRONMENTS ARE INCLUSIVE WITH NUEROD.',
    data=survey,
    palette='viridis'
)

plt.title("Comparative distribution between invitations and inclusive settings")
plt.xlabel("You invited someone with different behaviour at events")
plt.ylabel("Conunt")
plt.legend(title="Inclusive environments")
plt.show()
```

#### Relationship between people who have or have not invited possible neurodiverse people to social events and the lack of inclusive environments: indicates that most people prefer not to invite neurodiverse people to events, we can deduce from previous graphs that it is due to lack of information or knowing how to deal with them. Both possibilities of excluding or ajuntat coincide in that the environments are not inclusive or are in doubt.

### Conclusion

###### Based on the responses from 32 participants, the data reveals notable insights about social interactions with neurodiverse individuals:

Demographics: Slightly more women than men participated in the survey, and among the respondents, there are marginally more neurodivergent women than men.

Invitations and Comfort Levels: The number of individuals who invite neurodiverse people to social events is equal to those who do not. However, those who refrain from inviting neurodiverse individuals report feeling more discomfort in their presence. Interestingly, even among those who do extend invitations, a portion still experiences unease around people exhibiting atypical behavior.

Perceived Inclusivity and Measures: A significant majority of respondents believe that current social environments lack inclusivity for neurodiverse individuals. There is also a prominent sentiment that insufficient measures are in place to integrate these individuals, and many participants feel there is a lack of guidance, techniques, or resources to foster inclusivity.

Perception of Injustice: The survey highlights a widespread perception of social injustice against neurodiverse individuals. A clear pattern emerges, indicating that greater discomfort with neurodiverse individuals correlates with a heightened sense of perceived injustice toward them.

Knowledge and Support: While most respondents claim to offer help to neurodiverse individuals in their social environments, many admit to not knowing how to provide effective support. This points to a significant gap in knowledge and resources for fostering meaningful interactions.

Barriers to Inclusivity: The relationship between inviting neurodiverse individuals to social events and the lack of inclusive environments reveals a concerning trend. Many participants choose not to invite neurodiverse individuals, which, when analyzed alongside other data points, seems to stem from a lack of information or understanding about how to engage with them. Both those who include and those who exclude neurodiverse individuals agree that social spaces are not inclusive or are uncertain about their inclusivity.

General Observation:
The findings underscore the pressing need for increased awareness, education, and actionable measures to create genuinely inclusive social environments for neurodiverse individuals. Addressing the gaps in information and fostering a culture of understanding could significantly improve social interactions and reduce the perceived discomfort and injustice highlighted in this survey.
